```python
from openpyxl import Workbook
import pandas as pd

# Path to existing Excel file
file_path = r"C:\Users\SHRUTI S\Documents\Repos\Entab Internship\Data Analysis\ZeroData.xlsx"

# Path to new Excel file
output_file_path = r"C:\Users\SHRUTI S\Documents\Repos\Entab Internship\Data Analysis\ZeroDataWithKeywords.xlsx"

# Load data from Excel
data = pd.read_excel(file_path, sheet_name="Data")

# Define keywords
keywords = [
    'Updation', 'Assignment', 'Assigner', 'Reconciliation', 'module wise status',
    'Online Payment', 'Update', 'Paid', 'Promotion', 'Mark assigner', 'Design',
    'SMS', 'DLT', 'Training', 'DEMO', 'Concession', 'Fixed', 'Route Assigner',
    'Demote', 'Chargeback', 'Permission', 'Create', 'Preupgradation', 'Pre Upgradation',
    'Pre-upgradation', 'Upload', 'Entry', 'Fee structure', 'Fine Setting', 'Transfer',
    'Settings', 'Setting', 'collection', 'Adjustment', 'Creation', 'Create',
    'optional subject', 'optional', 'Modification', 'Fee Group', 'Assigning',
    'Tally', 'Tallied', 'visit', 'Data updation', 'Membership', 'Book Entry',
    'Bulk', 'Delete', 'Mapping'
]

# Initialize an empty list to store rows
rows = []

# Initialize an empty DataFrame for keywords (optional)
keywords_df = pd.DataFrame(columns=['Ticket ID', 'Subject', 'Description', 'Module'] + keywords)

# Create a new workbook
book = Workbook()

# Remove the default sheet created by openpyxl
book.remove(book.active)

# Create a sheet for writing the data
sheet = book.create_sheet('KeywordResults')

# Write headers
sheet.append(['Ticket ID', 'Subject', 'Description', 'Module'] + keywords)

# Write data
for idx, row in data.iterrows():
    token_id = row['Ticket ID']
    subject = row['Subject']
    description = str(row['Description'])
    module = row.get('Module', '')  # Get 'Module' column value or default to empty string if not present
    token_row = {'Ticket ID': token_id, 'Subject': subject, 'Description': description, 'Module': module}
    
    # Check for presence of keywords in description (case insensitive) and assign values
    for keyword in keywords:
        if keyword.lower() in description.lower():
            token_row[keyword] = 1
        else:
            token_row[keyword] = 0
    
    # Append current row to list of rows
    rows.append(token_row)

# Write data to the sheet
for row in rows:
    row_data = [row['Ticket ID'], row['Subject'], row['Description'], row['Module']] + [row[keyword] for keyword in keywords]
    sheet.append(row_data)

# Save workbook
book.save(output_file_path)

print(f"Keywords data has been successfully written to '{output_file_path}'")
```
